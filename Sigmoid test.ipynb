```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(x))
```

```python
x = range(1, 21)
y = []
for xx in x:
    y.append(sigmoid(xx/2.5 - 4) + 0.025)
    print xx, y[-1]
plt.plot(x, y)
```

```python
x = range(1, 21)
y = []
for xx in x:
    y.append((sigmoid(xx/1.5 - 7) + 0.053) * 0.91)
    print xx, y[-1]
plt.plot(x, y)
```
