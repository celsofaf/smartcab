```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(x))
```

```python
x = range(1, 101)
y = []
for xx in x:
    y.append(sigmoid(xx/12.5 - 4) * 0.98 + 0.032)
    print xx, y[-1]
plt.plot(x, y)
```
